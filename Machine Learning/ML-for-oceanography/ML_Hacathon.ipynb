```python
x=df.columns
plt.subplots(4,4,figsize=(18,18))
sum=1
for i in x:
  x1=df[i].values
  y=df['WVHT(m)'].values
  plt.subplot(4,4,sum)
  plt.scatter(x1,y,marker='.',color='g')
  plt.xlabel(i)
  plt.ylabel('WVHT(m)')
  plt.grid()
  sum+=1
plt.subplots_adjust(wspace=0.35,hspace=0.65)
plt.show()
```

by observing the data removing the columns year,month,day,hour,minutes,

```python
df=df.drop(['#YY','MM','DD','hh','mm'],axis=1)
df.head()
```

```python
df.boxplot(figsize=(14,7))
```

```python
df.plot.hist(figsize=(7,7))
```

```python
#Standariztion of the data
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
sc.fit(df.drop('WVHT(m)',axis=1))
fea=sc.transform(df.drop('WVHT(m)',axis=1))
scdata=pd.DataFrame(fea,columns = df.drop('WVHT(m)', axis=1).columns)
scdata
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components = 7)
scdata = pca.fit_transform(scdata)
```

```python
scdata=pd.DataFrame(scdata)
scdata
```

```python
from sklearn.model_selection import train_test_split
X=scdata
y=df['WVHT(m)'].values
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

```python
X
```

Gradient Boosting Regressor


```python
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.model_selection import train_test_split
import sklearn.metrics as metrics
# Train a gradient boosting regressor
gb_reg = GradientBoostingRegressor(n_estimators=200, learning_rate=0.27, max_depth=3, random_state=42)
gb_reg.fit(X_train, y_train)

# Evaluate the model on the test set
y_pred = gb_reg.predict(X_test)
error=np.sqrt(metrics.mean_squared_error(y_test, y_pred))
print(error)
```

```python
over=[]
o1=[]
under=[]
o2=[]
for i in range(len(y_test)):
  if(y_pred[i]>y_test[i]):
    over.append(y_pred[i])
    o1.append(y_test[i])
  elif(y_pred[i]<y_test[i]):
    under.append(y_pred[i])
    o2.append(y_test[i])
over=np.array(over)
under=np.array(under)
o1=np.array(o1)
o2=np.array(o2)
plt.scatter(o1,over,color='red')
plt.scatter(o2,under,color='green')
plt.plot(y_test,y_test,'-.',color='black')
plt.xlabel("Actual values")
plt.ylabel("Predicted values")
plt.grid()
plt.show()
```

ANN model


```python
from keras.models import Sequential
from keras.layers import Dense

# define the input shape
input_shape = (7,)

# create the model
model = Sequential()
model.add(Dense(64, activation='relu', input_shape=input_shape))
model.add(Dense(64, activation='relu'))
model.add(Dense(1, activation='linear'))

# compile the model
model.compile(loss='mse', optimizer='adam')

# fit the model on training data
model.fit(X_train, y_train, epochs=30, batch_size=32)

# evaluate the model on test data
loss = model.evaluate(X_test, y_test)
```
