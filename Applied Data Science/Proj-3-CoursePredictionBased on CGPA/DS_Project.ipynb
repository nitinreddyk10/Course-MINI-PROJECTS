**Loading the data**

```python
from google.colab import drive
drive.mount('/content/drive')
df=pd.read_csv("/content/drive/MyDrive/ds.csv")
#df=pd.read_csv("/content/drive/MyDrive/DS/DS_Project.csv")
df
```

```python
replacements={
    '>=9.0':'9',
    '>=8.0':'8',
    '>=7.0':'7',
    '>=6.0':'6',
    '<6':'5',
}
df.replace(replacements,inplace=True)
df['Elective']=None
df['Elective']=df['Professional Elective'].fillna(df['Professional Elective.1']).fillna(df['Professional Elective.2'])
df = df.drop(['Professional Elective', 'Professional Elective.1', 'Professional Elective.2'], axis=1)
df
```

Step-1 : DATA CLEANING

Handling Duplicates:

```python
#drop rows with null values
df.dropna(subset=['Elective'], inplace=True)
df.drop_duplicates(subset=['Email Address'], keep='first', inplace=True)
# drop unnecessary columns
df.drop(['Timestamp', 'Email Address'], axis=1, inplace=True)
df
```

```python
# Split the 'Area of Interest' column by comma and convert to a list
df['Area of interest'] = df['Area of interest'].str.split(',')
# Split the 'final year project major fields' column by comma and convert to a list
df['Project Details'] = df['Project Details'].str.split(',')
# Split the 'Elective' column by comma and convert to a list
df['Elective'] = df['Elective'].str.split(',')
df
```

Handling Missing Values:

```python
# Convert the CGPA column to float
df['Cgpa'] = pd.to_numeric(df['Cgpa'])
# Fill missing values in the CGPA column with the median CGPA
df['Cgpa'].fillna(df['Cgpa'].median(), inplace=True)
# Fill missing values in the Area of interest  column with the mode
df['Area of interest'].fillna(df['Area of interest'].mode(), inplace=True)
# Fill missing values in the Elective column with the mode
df['Elective'].fillna(df['Elective'].mode(), inplace=True)

df.dropna(subset=['Project Details'], inplace=True)
df
```

```python
df['Area of interest'] = df['Area of interest'].apply(lambda x: x[0] if isinstance(x, list) and len(x) > 0 else x)
df['Project Details'] = df['Project Details'].apply(lambda x: x[0] if isinstance(x, list) and len(x) > 0 else x)
df['Elective'] = df['Elective'].apply(lambda x: x[0] if isinstance(x, list) and len(x) > 0 else x)
df
```

```python
# Convert all items in 'Area of interest' column to uppercase
df['Area of interest'] = df['Area of interest'].apply(lambda x: x.upper() if isinstance(x, str) else x)
# Convert all items in 'Project Details' column to uppercase
df['Project Details'] = df['Project Details'].apply(lambda x: x.upper() if isinstance(x, str) else x)
df['Branch'] = df['Branch'].apply(lambda x: x.upper() if isinstance(x, str) else x)
df
```

Handling outliers

```python
replacements_1={
    'DS':'1',
    'ML':'2',
    'AI':'3',
    'WEB DEVELOPMENT':'4',
    'CV':'5',
    'EMBEDDED SYSTEMS':'6',
    'VLSI':'7',
    'IIOT':'8',
    '3D MODELLING':'9',
    'OTHERS':'10',
}
replacements_2={
    'Pattern Recognition and Machine Learning':'1',
    'Digital Image Processing(DIP)':'2',
    'Biometrics':'3',
    'IIOT':'4',
    'Introduction to Photonics':'5',
    'Analog Design':'6',
    'Design for Artificial Organ':'7',
    'Medical Image Analysis':'8',
    'Principles of Vibration':'9',
    'Design of machine elements':'10',
    'computational fluid dynamics':'11',
    'Design for manufacturing and automation':'12',
}
replacements_3={
    'CSE':'1',
    'ECE':'2',
    'ME':'3',
    'SM':'4',
    'OTHERS':'5',
}
df['Project Details'] = df['Project Details'].replace(replacements_1)
df['Area of interest'] = df['Area of interest'].replace(replacements_1)
df['Elective'] = df['Elective'].replace(replacements_2)
df['Branch'] = df['Branch'].replace(replacements_3)
df
```
